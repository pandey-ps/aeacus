```python
import scanpy as sc
```

```python
FILE = 'tsep.h5ad'
```

```python
adata = sc.read_h5ad(FILE)
adata.var_names_make_unique()
print(f'{adata.n_obs:,} cells x {adata.n_vars:,} genes')
if 'cell_type' in adata.obs.columns:
    print(adata.obs['cell_type'].value_counts().head(8).to_string())
```

```python
from aeacus import Profiler
result   = Profiler(test_input=adata.copy(), norm_type='cpm_log1p', use_raw = True).load().profile() # data given is in raw format
n_normal = (result.obs['malignancy_call'] == 'Normal').sum()
n        = len(result)
print(f'Predicted Normal    : {n_normal} / {n}')
print(f'Predicted Malignant : {n - n_normal} / {n}')
print(f'Accuracy (specificity): {n_normal / n:.2%}')
```
