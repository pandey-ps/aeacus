```python
import urllib.request, os
import numpy as np, pandas as pd, scanpy as sc, scipy.sparse as sp
```

```python
URL  = "https://ftp.ncbi.nlm.nih.gov/geo/series/GSE81nnn/GSE81383/suppl/GSE81383_data_melanoma_scRNAseq_BT_Mel.txt.gz"
FILE = "GSE81383_melanoma.txt.gz"
```

```python
if not os.path.exists(FILE):
    urllib.request.urlretrieve(URL, FILE)
```

```python
df   = pd.read_csv(FILE, sep="\t", index_col=0)
adata = sc.AnnData(
    X   = sp.csr_matrix(df.values.T.astype(np.float32)),
    obs = pd.DataFrame(index=df.columns),
    var = pd.DataFrame(index=df.index),
)
adata.var_names_make_unique()
f"{adata.n_obs} cells × {adata.n_vars} genes"
```

```python
# normalizing data
from aeacus import Profiler
result = Profiler(test_input=adata.copy(), norm_type="tpm_log1p").load().profile() # raw = False (default) as data is already normalized
n_correct = (result.obs["malignancy_call"] == "Malignant").sum()
accuracy  = n_correct / len(result)
accuracy
```
